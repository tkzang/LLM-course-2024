```python
# Initialize LLm
llm = Ollama(model="llama3", request_timeout=60.0)

# Define URLs and Reader
llmsherpa_api_url = "http://localhost:5010/api/parseDocument?renderFormat=all"
pdf_url = "https://abc.xyz/assets/91/b3/3f9213d14ce3ae27e1038e01a0e0/2024q1-alphabet-earnings-release-pdf.pdf"
pdf_reader = LayoutPDFReader(llmsherpa_api_url)

# Read PDF
doc = pdf_reader.read_pdf(pdf_url)

# Load all sections and search for relevant data
context = ""
for section in doc.sections():
    # Dynamically add all sections to the context
    context += section.to_html(include_children=True, recurse=True)


questions = [
    "What was Google's operating margin for 2024?",
    "What % Net income is of the Revenues?"
]

# Query the LLM for each question
for question in questions:
    resp = llm.complete(f"read this table and answer question: {question}:\n{relevant_context}")
    print(resp.text)
```
